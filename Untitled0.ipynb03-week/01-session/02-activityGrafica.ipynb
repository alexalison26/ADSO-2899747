```python
# ==============================
#  Importar libreras
# ==============================
from google.colab import files
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ==============================
#  Subir y cargar el archivo CSV
# ==============================
uploaded = files.upload()  # Te pedir seleccionar el archivo desde tu PC

df = pd.read_csv("bengaluru_house_prices.csv")
print("Primeras filas del dataset:")
print(df.head())

# Extraer nmero de habitaciones de la columna "size"
df['num_habitaciones'] = df['size'].str.extract('(\d+)').astype(float)

# ==============================
#  1. Grfico de barras - Precio promedio segn tipo de rea
# ==============================
plt.figure(figsize=(8,5))
sns.barplot(x="area_type", y="price", data=df, estimator="mean", palette="Set2")
plt.title("Precio promedio de vivienda segn tipo de rea")
plt.xlabel("Tipo de rea")
plt.ylabel("Precio promedio (lakh INR)")
plt.xticks(rotation=30)
plt.show()

# ==============================
#  2. Grfico de torta - Distribucin de viviendas por tipo de rea
# ==============================
plt.figure(figsize=(6,6))
df['area_type'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set2"))
plt.title("Distribucin de viviendas segn tipo de rea")
plt.ylabel("")
plt.show()

# ==============================
#  3. Grfico de barras - Promedio de baos por nmero de habitaciones
# ==============================
plt.figure(figsize=(8,5))
sns.barplot(x="num_habitaciones", y="bath", data=df, estimator="mean", palette="Set3")
plt.title("Promedio de baos segn nmero de habitaciones")
plt.xlabel("Nmero de habitaciones")
plt.ylabel("Promedio de baos")
plt.show()

# ==============================
#  4. Grfico de barras - Precio promedio segn nmero de habitaciones
# ==============================
plt.figure(figsize=(8,5))
sns.barplot(x="num_habitaciones", y="price", data=df, estimator="mean", palette="Set1")
plt.title("Precio promedio segn nmero de habitaciones")
plt.xlabel("Nmero de habitaciones")
plt.ylabel("Precio promedio (lakh INR)")
plt.show()

```
