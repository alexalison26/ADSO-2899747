```python
# ==============================
# 📌 Importar librerías
# ==============================
from google.colab import files
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ==============================
# 📌 Subir y cargar el archivo CSV
# ==============================
uploaded = files.upload()  # Te pedirá seleccionar el archivo desde tu PC

df = pd.read_csv("bengaluru_house_prices.csv")
print("Primeras filas del dataset:")
print(df.head())

# Extraer número de habitaciones de la columna "size"
df['num_habitaciones'] = df['size'].str.extract('(\d+)').astype(float)

# ==============================
# 📊 1. Gráfico de barras - Precio promedio según tipo de área
# ==============================
plt.figure(figsize=(8,5))
sns.barplot(x="area_type", y="price", data=df, estimator="mean", palette="Set2")
plt.title("Precio promedio de vivienda según tipo de área")
plt.xlabel("Tipo de área")
plt.ylabel("Precio promedio (lakh INR)")
plt.xticks(rotation=30)
plt.show()

# ==============================
# 🥧 2. Gráfico de torta - Distribución de viviendas por tipo de área
# ==============================
plt.figure(figsize=(6,6))
df['area_type'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set2"))
plt.title("Distribución de viviendas según tipo de área")
plt.ylabel("")
plt.show()

# ==============================
# 📊 3. Gráfico de barras - Promedio de baños por número de habitaciones
# ==============================
plt.figure(figsize=(8,5))
sns.barplot(x="num_habitaciones", y="bath", data=df, estimator="mean", palette="Set3")
plt.title("Promedio de baños según número de habitaciones")
plt.xlabel("Número de habitaciones")
plt.ylabel("Promedio de baños")
plt.show()

# ==============================
# 📊 4. Gráfico de barras - Precio promedio según número de habitaciones
# ==============================
plt.figure(figsize=(8,5))
sns.barplot(x="num_habitaciones", y="price", data=df, estimator="mean", palette="Set1")
plt.title("Precio promedio según número de habitaciones")
plt.xlabel("Número de habitaciones")
plt.ylabel("Precio promedio (lakh INR)")
plt.show()

```
